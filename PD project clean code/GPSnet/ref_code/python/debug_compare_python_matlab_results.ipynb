```python
import numpy as np
import pandas as pd
import os
```

```python
python_result_path='../../FMNW_interpolation/python/Final_Modules/pval0.05/'
matlab_result_path='../../FMNW_interpolation/matlab/pval0.05/GPSnet_result/'
cutoff_list=[0.005,0.05,0.01,0.5]
file_list=os.listdir(python_result_path)
file_list=[i for i in file_list if i.endswith('_0.005.txt')]
trait_list=[('_').join(i.split('_')[:-1]) for i in file_list]

stats_cols=['trait']
for cutoff in cutoff_list:
    prefix='cutoff'+str(cutoff)
    stats_cols+=[prefix+'_pyNum',prefix+'_matNum',prefix+'_overlap',prefix+'_jaccard']
summary_df=pd.DataFrame(columns=stats_cols)

for trait in trait_list:
    new_row_list=[trait]
    for cutoff in cutoff_list:
        f_py=open(python_result_path+trait+'_'+str(cutoff)+'.txt','r')
        f_mat=open(matlab_result_path+trait+'_FamaMacBeth_NeweyWest_cutoff_'+str(cutoff)+'_alpha_0.5.txt','r')
        list_py=f_py.readlines()
        list1=[int(i.rstrip()) for i in list_py[1:]]#first element is 'gene'
        list_mat=f_mat.readlines()
        list2=[int(i.rstrip()) for i in list_mat[1:]]
        
        new_row_list.append(len(list1))
        new_row_list.append(len(list2))

        array1 = np.array(list1)
        array2 = np.array(list2)
        overlap_count = len(np.intersect1d(array1, array2))
        new_row_list.append(overlap_count)

        jaccard=overlap_count/len(set(list1+list2))
        new_row_list.append(jaccard)

    new_row_df = pd.DataFrame([new_row_list], columns=stats_cols)
    summary_df = summary_df.append(new_row_df, ignore_index=True)

```

```python
summary_df.to_csv('../summary_stats_pyVSmatlab.csv',index=False)
```
